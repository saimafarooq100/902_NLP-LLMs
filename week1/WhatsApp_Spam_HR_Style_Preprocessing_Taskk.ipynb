## Step 1: Load the dataset

```
import pandas as pd
url = "https://raw.githubusercontent.com/mohitgupta-omg/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv"
messages_df = pd.read_csv(url, encoding="latin-1")
messages_df.head()
```

## Step 2: Keep only useful columns

```
messages_df = messages_df[['v1', 'v2']]
messages_df.columns = ['message_type', 'message_text']
messages_df.head()
```

## Step 3: Spam vs Ham count

```
messages_df['message_type'].value_counts()
```

## Step 4: Calculate message length

```
lengths = []
for text in messages_df['message_text']:
    lengths.append(len(text))
messages_df['message_length'] = lengths
messages_df[['message_type','message_length']].head()
```

## Step 5: Convert text to lowercase

```
lower_text = []
for text in messages_df['message_text']:
    lower_text.append(text.lower())
messages_df['clean_text'] = lower_text
```

## Step 6: Remove URLs

```
import re
text_no_url = []
for text in messages_df['clean_text']:
    text_no_url.append(re.sub(r'https?://\S+|www\.\S+', '', text))
messages_df['clean_text'] = text_no_url
```

## Step 7: Remove phone numbers

```
text_no_numbers = []
for text in messages_df['clean_text']:
    text_no_numbers.append(re.sub(r'[\+]?[\d\s\-]{7,}', '', text))
messages_df['clean_text'] = text_no_numbers
```

## Step 8: Remove punctuation

```
import string
text_no_punct = []
for text in messages_df['clean_text']:
    text_no_punct.append(text.translate(str.maketrans('', '', string.punctuation)))
messages_df['clean_text'] = text_no_punct
```

## Step 9: Normalize spaces

```
final_space = []
for text in messages_df['clean_text']:
    final_space.append(re.sub(r'\s+', ' ', text).strip())
messages_df['clean_text'] = final_space
```

## Step 10: Tokenisation

```
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
tokens = []
for text in messages_df['clean_text']:
    tokens.append(word_tokenize(text))
messages_df['tokens'] = tokens
```

## Step 11: Remove stopwords

```
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
clean_tokens = []
for token_list in messages_df['tokens']:
    kept = []
    for word in token_list:
        if word not in stop_words:
            kept.append(word)
    clean_tokens.append(kept)
messages_df['tokens'] = clean_tokens
```

## Step 12: Remove short words

```
final_tokens = []
for token_list in messages_df['tokens']:
    kept = []
    for word in token_list:
        if len(word) >= 2:
            kept.append(word)
    final_tokens.append(kept)
messages_df['tokens'] = final_tokens
```

## Step 13: Final clean message

```
final_text = []
for token_list in messages_df['tokens']:
    final_text.append(' '.join(token_list))
messages_df['final_clean_text'] = final_text
messages_df[['message_type','final_clean_text']].head()
```

## Step 16: Spam vs Ham distribution

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=messages_df, x='message_type')
plt.title('Spam vs Ham Messages')
plt.show()
```

## Step 17: Message length comparison

```
sns.boxplot(data=messages_df, x='message_type', y='message_length')
plt.title('Message Length Comparison')
plt.show()
```
